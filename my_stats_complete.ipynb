```python
data = sns.load_dataset('tips')
data
```

```python
data[["sex", "smoker"]]
```

```python
data_table = pd.crosstab(data['sex'], data['smoker'])
data_table
```

```python
pd.crosstab(data['sex'], data['smoker'])
```

```python
observed_values = data_table.values
observed_values
```

```python
import scipy.stats as stats
```

```python
stats_test,p,dof,Expected_Values=stats.chi2_contingency(observed_values)
```

```python
Expected_Values
```

```python
no_of_rows = data_table.iloc[0:2, 0]
no_of_rows
```

```python
no_of_columns = data_table.iloc[0, 0:2]
no_of_columns
```

```python
dof=(len(no_of_columns)-1)*(len(no_of_rows)-1)
```

```python
dof
```

```python
observed_values
```

```python
Expected_Values
```

```python
for o, e in zip(observed_values, Expected_Values):
    print(o, e)
```

```python
from scipy.stats import chi2
chisqure_test=sum([(o-e)**2./e for o,e in zip(observed_values,Expected_Values)])
```

```python
chisqure_test
```

```python

chi_square_statistic=chisqure_test[0]+chisqure_test[1]
chi_square_statistic
```

```python
dof
```

```python
alpha = 0.05
```

```python
from scipy.stats import chi2
critical_value=chi2.ppf(1-alpha,df=dof)
```

```python
critical_value
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=dof)
```

```python
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
# no if hours student study in a weekly basis daily
expeted_data=[8,6,7,9,6,9,7]
observed_data=[7,8,6,9,9,6,7]
```

```python
sum(expeted_data)
```

```python
sum(observed_data)
```

```python
## chi square -- goodnes of fit Test
chisquare_test_statistic,p_value=stat.chisquare(observed_data,expeted_data)
```

```python
chisquare_test_statistic,p_value
```

```python
## find the critical values
signifiance_value=0.05
dof=len(expeted_data)-1
print(dof)
critical_value=stat.chi2.ppf(1-signifiance_value,dof)
```

```python
critical_value
```

```python
if chisquare_test_statistic> critical_value:
    print("reject the null hypothesis")
else:
    print("we accepted the null hypothises")
```

```python
# Evudence to say that two population variances are not equal
worker1=[18,19,22,25,27,28,41,45,51,55]
worker2=[14,15,15,17,18,22,25,25,27,34]
```

```python
import numpy as np
```

```python
## calculating f test
f_test=np.var(worker1)/np.var(worker2)
```

```python
f_test
```

```python
## degree of freedom
df1=len(worker1)-1
df2=len(worker2)-1
signifiance_value=0.05
```

```python
df1
```

```python
df2
```

```python
import scipy.stats as stat
```

```python
critical_value=stat.f.ppf(q=1-signifiance_value,dfn=df1,dfd=df2)
```

```python
critical_value
```

```python
if f_test>critical_value:
    print("Reject the null hypothesis")
else:
    print("we faill to rejeact the null hypothesis")
```
